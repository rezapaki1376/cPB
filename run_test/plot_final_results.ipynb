```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1) plot CPB accuracy and cohen kappa

```python
df1 = np.array(pd.read_csv('results/acc-CPB-16_16-1432-t1.csv'))
df2 = np.array(pd.read_csv('results/acc-CPB-16_16-1432-t2.csv'))
df3 = np.array(pd.read_csv('results/acc-CPB-16_16-1432-t3.csv'))
df4 = np.array(pd.read_csv('results/acc-CPB-16_16-1432-t4.csv'))
all_x=np.concatenate((df1,df2,df3,df4),axis=0)
y = np.cumsum(all_x, axis=1) / np.arange(1, all_x.shape[1] + 1)
vertical_lines_x = [all_x.shape[1], 2*all_x.shape[1], 3*all_x.shape[1]]
x = list(range(1,4*all_x.shape[1] + 1))
fig, ax = plt.subplots(figsize=(12, 4))
ax.margins(x=0.0)
label_list=['CPB Task 1','CPB Task 2','CPB Task 3','CPB Task 4']

for i in vertical_lines_x:
  plt.axvline(x=i, color='#D3D3D3', linestyle='-')

for i in range(0,y.shape[0],4):
  ax.plot(x, np.concatenate((y[i],y[i+1],y[i+2],y[i+3]),axis=0), label= label_list[int(i/4)])
plt.xlabel('Batch Number')
plt.ylabel('Cumulative Accuracy')
plt.title('Cumulative Accuracy Over Batches on Dataset 1432')
plt.legend()
plt.show()
```
